# I. A Simple Linear Regression Model

$$y = a_0 + a_1 x + \epsilon$$

````python
from sklearn.linear_model import LinearRegression
````

`Initialize an instance of linear regression:`

```python
linear_model = LinearRegression(fit_intercept=True, 
                                normalize=False, 
                                copy_X=True)
```

* `fit_intercept:` if `True` the intercept will be part of the model
* `normalize:` converts an input value to a boolean
* `copy_X:` if `True` X will be copied, otherwise it may be overwritten

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

````python
# Create a numpy array of data:
x = np.array([6, 16, 26, 36, 46, 56]).reshape((-1, 1))
y = np.array([4, 23, 10, 12, 22, 35])
````

````python
x, y
````

````python
fig1, ax1 = plt.subplots(1,1, figsize=(7,5))
ax1.scatter(x, y, label='data')
````

````python
# Create an instance of a linear regression model: 
l_model = LinearRegression()
````

````python
# Fit the model to the data with the fit() function:
l_model.fit(x, y)
````

````python
# look at the coefficient values for the INTERCEPT of the linear regression model:
a0 = l_model.intercept_
print('Intercept:', a0) 
````

````python
# look at the coefficient value for the SLOPE of the linear regression model:
a1 = l_model.coef_
print('Slope:', a1)
````

````python
# Obtain the coefficient of determination by calling the model with the score() function:
R2 = l_model.score(x, y)
print('Coefficient of determination:', R2)
````

````python
# Predict the model's response at the given 'x' and print it:
y_pred = l_model.predict(x)
print('Predicted response:', y_pred, sep='\n')
````

````python
l_model.get_params()
````

### Other metrics:

````python
from sklearn.metrics import mean_absolute_error
````

* expresses the average absolute model prediction error
* a perfect model would score 0
* this metric has no upper bounds
* it is measured in the units of the target, thus it is more interpretable
* a good metric if you want to ignore the outliers
* cannot indicate how bad a model is
* can be used to compare two models: the lower the better

````python
MAE = mean_absolute_error(y, y_pred)
print('MAE:', MAE)
````

````python
from sklearn.metrics import mean_squared_error
````

* it penalizes large errors
* cannot indicate how bad a model is
* can be used to compare two models: the lower the better

````python
MSE = mean_squared_error(y, y_pred)
print('MSE:', MSE)
````

````python
from sklearn.metrics import r2_score
````

* it takes values between 0 and 1
* higher values are better, however
* it is difficult to evaluate the model from this metric alone: is 0.7 a good number?
* use this metric in combination with other metrics or visualizations

````python
r2 = r2_score(y, y_pred)
print('Coefficient of determination:', r2)
````

````python
ypred = lambda x: a0 + a1*x
````

````python
fig1, ax1 = plt.subplots(1,1, figsize=(7,5))
ax1.scatter(x, y, label='data')
ax1.plot(x, ypred(x), label='linear fit')
ax1.legend()
ax1.set_xlabel('x')
ax1.set_ylabel('y')
````

# II. Polynomial Regression

`sklearn.preprocessing.PolynomialFeatures`

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html

$$y = a_0 + a_1 x + a_2 x^2 + \cdots + a_n x^n$$

````python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
````

````python
fig2, ax2 = plt.subplots(1,1, figsize=(9,5))
ax2.scatter(X, y, label='data')
````

````python
from sklearn.preprocessing import PolynomialFeatures
````

````python
Degree = 2
poly_features = PolynomialFeatures(degree=Degree, include_bias=False)
````

````python
X_poly = poly_features.fit_transform(X)
````

````python
poly_features.get_feature_names_out( ['x'] )
````

````python
# Create an instance of a quadratic regression model: 
q_model = LinearRegression()
````

````python
# Fit the model to the data with the fit() function:
q_model.fit(X_poly, y)
````

````python
# look at the coefficient values for the INTERCEPT of the quadratic regression model:
b0 = q_model.intercept_
print('Intercept:', b0)
````

````python
# look at the coefficient value for the SLOPE of the quadratic regression model:
b1 = q_model.coef_
print('Slope:', b1)
````

````python
y2pred = lambda x: b1[0,1]*x**2 + b1[0,0]*x + b0
````

````python
xpred = np.linspace(min(X), max(X))
fig2, ax2 = plt.subplots(1,1, figsize=(9,5))
ax2.scatter(X, y, label='data')
ax2.plot(xpred, y2pred(xpred), color='red', label='quadratic fit')
ax2.legend()
ax2.set_xlabel('x')
ax2.set_ylabel('y')
````

````python
sns.residplot(x=X, y=y, order=2, label='residuals for quadratic fit')
````

````python
sns.residplot(x=X, y=y, order=1, label='residuals for linear fit')
````

````python
res = y - y2pred(X)
````

````python
plt.scatter(y2pred(X), res)
plt.xlabel(r'$\hat{y}$')
plt.ylabel('residuals')
````

# Lasso Regression

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html#sklearn.linear_model.LassoCV

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.lasso_path.html#sklearn.linear_model.lasso_path

https://scikit-learn.org/stable/auto_examples/linear_model/plot_lasso_coordinate_descent_path.html#sphx-glr-auto-examples-linear-model-plot-lasso-coordinate-descent-path-py

$$C = \sum_{i=1}^{n} \left({y_i - \sum_{j=0}^{m} 	\beta_j f_{ji}}\right)^2 + \alpha \sum_{i=1}^{n} \lvert\beta_i \rvert$$

````python
from sklearn.linear_model import Lasso
````

````python
lasso_model = Lasso(alpha=0.1,
                    copy_X=True,
                    fit_intercept=True,
                    random_state=5)
````

**Returns:**
* alpha_: The amount of penalization chosen by cross validation
* alphas_: The grid of alphas used for fitting
* coef_: Parameter vector
* intercept_: Independent term in decision function
* mse_path_: Mean square error for the test set on each fold, varying alpha

````python
lasso_model.fit(X_train, y_train)
````

````python
print(lasso_model.coef_)
````

````python
print(lasso_model.intercept_)
````

````python
print(lasso_model.n_features_in_)
````

````python
print('Coefficient of determination - training set:', lasso_model.score(X_train, y_train))
````

````python
print('Coefficient of determination - testing set:', lasso_model.score(X_test, y_test))
````

# Ridge

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html

https://scikit-learn.org/stable/auto_examples/linear_model/plot_ridge_path.html#sphx-glr-auto-examples-linear-model-plot-ridge-path-py

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeCV.html#sklearn.linear_model.RidgeCV

$$C = \sum_{i=1}^{n} \left({y_i - \sum_{j=0}^{m} 	\beta_j f_{ji}}\right)^2 + \alpha \sum_{i=1}^{n} \beta_i^2$$

* When alpha is very large, the regularization effect dominates the squared loss function and the coefficients tend to zero.
* As alpha tends towards zero, the solution tends towards the ordinary least squares solution.

````python
from sklearn.linear_model import Ridge
````

````python
# Compute paths

n_alphas = 200
alphas = np.logspace(-3, 6, n_alphas)

coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=True)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)

# #############################################################################
# Display results

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale("log")
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Ridge coefficients as a function of the regularization")
plt.axis("tight")
plt.show()
````
